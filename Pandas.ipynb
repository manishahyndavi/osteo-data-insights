```python
# 2 sample t test

from scipy.stats import ttest_ind
females = final[final['Gender'] == 'F']['DEXA']
males = final[final['Gender'] == 'M']['DEXA']

res = ttest_ind(females,males)
print(res)
```

```python
count = final.groupby(['Race_name']).groups
print(count)
```

```python
# one way anova

from scipy.stats import f_oneway

african = final[final['Race_name'] == 'African'] ['Vit D (ng/ml)']
asian = final[final['Race_name'] == 'Asian'] ['Vit D (ng/ml)']
his = final[final['Race_name'] == 'Hispanic'] ['Vit D (ng/ml)']
nhis = final[final['Race_name'] == 'Non-Hispanic'] ['Vit D (ng/ml)']
mex = final[final['Race_name'] == 'Mexican American'] ['Vit D (ng/ml)']

res1 = f_oneway(african,asian,his,nhis,mex)

print(res1)
```

```python
# one sample ttest

from scipy.stats import ttest_1samp
 
res2 = ttest_1samp(final['DEXA'],-0.6)

print(res2)
```
